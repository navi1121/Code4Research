```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('/content/drive/MyDrive/Colab Notebooks/data/N2V/png_images/longbeach.png')
img_np = np.array(img)/255
x,y,channels = img_np.shape
plt.imshow(img_np) #inital psnr
#final psnr
#effect of patch size on psnr
#role of training and validation
#structure of neural network

```

```python
# Set the mean and standard deviation of the Gaussian distribution
mean = 0
std_dev = 0.15 #adjust as needed

noise = np.random.normal(loc = mean, scale = std_dev, size = (x,y,1))
#noise = np.clip(noise, 0, 1)
plt.imshow(noise, cmap = 'gray')
plt.show()
```

```python
noisy = img_np + noise
noisy = np.clip(noisy, 0, 1)
plt.imshow(noisy)
```

```python
drive_path = '/content/drive/MyDrive/Colab Notebooks/data/N2V/png_images/Noisy'
file_name = 'longbeachNoise.png'
drive_path = drive_path + '/' + file_name
im = Image.fromarray(np.uint8(noisy*255))
im.save(drive_path)
```
