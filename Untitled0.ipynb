```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.image import imread, imsave
import cv2
from google.colab.patches import cv2_imshow
```

```python
def PSNR(gt, img):
    mse = np.mean(np.square(gt - img))
    return 20 * np.log10(255) - 10 * np.log10(mse)
```

```python
pred = cv2.imread('file (1) (1).png')
gt = cv2.imread('longbeach.png')
```

```python
cv2_imshow(pred)
cv2_imshow(gt)
```

```python
value = PSNR(gt, pred)
print(value)
```

```python
noisy = cv2.imread('file (1).png')
cv2_imshow(noisy)
```

```python
value2 = PSNR(noisy,pred)
print(value2)
```
